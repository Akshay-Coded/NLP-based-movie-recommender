The data is analyzed and the nature of the data is understood.Then the data is cleaned and pre-processed to implement machine learning model

#### Merging two Datasets

```python
movies = movies.merge(credits, on='title')
```

#### Dropping irrelevant columns

```python
movies = movies[['movie_id', 'title','overview', 'genres','keywords','cast','crew']]
```

```python
movies.head()
```

#### Removing rows that has null values

```python
movies.isnull().sum()
```

```python
movies.dropna(inplace=True)
```

```python
movies.isnull().sum()
```

#### Checking for duplicates

```python
movies.duplicated().sum()
```

#### Extracting Genres, keywords, cast and crew

```python
import ast

def convert(text):
    l=[]
    for i in ast.literal_eval(text):
        l.append(i['name'])
    return l
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
movies.head()
```

```python
def convert_cast(text):
    l=[]
    cnt=0
    for i in ast.literal_eval(text):
        l.append(i['name'])
        cnt += 1
        if cnt == 3:
            break
    return l
```

```python
movies['cast'] = movies['cast'].apply(convert_cast)
```

```python
movies.head()
```

```python
def get_dir(text):
    l=[]
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            l.append(i['name'])
            break
    return l
```

```python
movies['crew'] = movies['crew'].apply(get_dir)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
movies.head()
```

#### Removing spaces to avoid problem while using ML model

```python
def remove_space(L):
    L1 = []
    for i in L:
        L1.append(i.replace(" ",""))
    return L1
```

```python
movies['cast'] = movies['cast'].apply(remove_space)
movies['crew'] = movies['crew'].apply(remove_space)
movies['genres'] = movies['genres'].apply(remove_space)
movies['keywords'] = movies['keywords'].apply(remove_space)
```

```python
movies.head()
```

#### Concatinating all columns into tags to implement the model

```python
movies['tags'] = movies['overview'] + movies['genres'] + movies['keywords'] + movies['cast'] + movies['crew']
```

```python
movies.head()
```

```python
movie_upd = movies[['movie_id','title','tags']]
```

```python
movie_upd.head()
```

#### Converting tags as list to string

```python
movie_upd['tags'] = movie_upd['tags'].apply(lambda x: " ".join(x))
movie_upd.head()
```

```python
movie_upd['tags'] = movie_upd['tags'].apply(lambda x:x.lower())
movie_upd.head()
```

## Building Machine Learning Model

```python
import nltk
from nltk.stem import PorterStemmer
```

We use stemming to remove the suffixes -er, -ed, -ing, -ly, -ful, able, -s and -es etc.

```python
ps = PorterStemmer()
```

```python
def stems(text):
    T = []
    
    for i in text.split():
        T.append(ps.stem(i))
    
    return " ".join(T)
```

```python
movie_upd['tags'] = movie_upd['tags'].apply(stems)
```

```python
movie_upd.iloc[0]['tags']
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
vector = cv.fit_transform(movie_upd['tags']).toarray()
```

```python
vector
```

```python
vector.shape
```

```python
from sklearn.metrics.pairwise import cosine_similarity
similarity = cosine_similarity(vector)
similarity
```

```python
movie_upd[movie_upd['title'] == 'Spectre'].index[0]
```

```python
def recommend(movie):
    index = movie_upd[movie_upd['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(movie_upd.iloc[i[0]].title)
```

```python
recommend('Avatar')
```

```python
recommend('Hulk')
```

```python
recommend('Cars 2')
```
